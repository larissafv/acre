```python
import geobr
import pandas as pd
import geopandas as gpd
import random
from shapely.geometry import Point, Polygon
```

```python
#Base de dados de população por município
pop = pd.read_csv("data/2019.csv", sep = ";")
pop['GEOCODIGO'] = pop['COD_UF']*100000 + pop['COD_MN']

#Codigo de 6 dígitos
trans_cod = pd.read_csv("data/cod6.csv", sep = ";")
pop = pd.merge(pop, trans_cod, on = 'GEOCODIGO')

pop
```

```python
#Estados
est = pop.groupby(['COD_UF']).sum()
est = est.iloc[:,[1]]
est.to_csv("results/Estados.csv", sep = ";")
est
```

```python
#Microrregioes
micro = pd.read_csv("data/cod_mic.csv", sep = ";")
micro = pd.merge(pop, micro, on = "Codmun6")
micro = micro.groupby(['COD_MIC']).sum()
micro = micro.iloc[:, [2]]
micro.to_csv("results/Microrregioes.csv", sep = ";")
micro
```

```python
#Mesorregioes
meso = pd.read_csv("data/cod_meso.csv", sep = ";")
meso = pd.merge(pop, meso, on = "GEOCODIGO")
meso = meso.groupby(['COD_MESO']).sum()
meso = meso.iloc[:,[2]]
meso.to_csv("results/Mesorregioes.csv", sep = ";")
meso
```

```python
#Grandes Regiões
grand_reg = pd.read_csv("data/cod_reg.csv", sep = ";")
grand_reg = pd.merge(est, grand_reg, on = "COD_UF")
grand_reg = grand_reg.groupby(['COD_REG']).sum()
grand_reg = grand_reg.iloc[:,[1]]
grand_reg.to_csv("results/GrandesRegioes.csv", sep = ";")
grand_reg
```

```python
#Regioes Metropolitanas
reg_met = pd.read_csv("data/cod_met.csv", sep = ";")
reg_met = pd.merge(pop, reg_met, on = "Codmun6")
reg_met = reg_met.groupby(['COD_MET']).sum()
reg_met = reg_met.iloc[:,[2]]
reg_met.to_csv("results/RegioesMetropolitanas.csv", sep = ";")
reg_met
```

```python
#Regioes Intermediarias
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="CDA UFMG")
```

```python
inter_shp = gpd.read_file("data/reg_int/BR_RG_Intermediarias_2019.shp")
inter_shp
```

```python
mun = geobr.read_municipality(year = 2018)
mun = pd.merge(mun, pop, left_on = "code_muni", right_on = "GEOCODIGO")
mun
```

```python
points = []
for polygon in mun.geometry:
    minx, miny, maxx, maxy = polygon.bounds
    b = False
    while(b == False):
        pnt = Point(random.uniform(minx, maxx), random.uniform(miny, maxy))
        if polygon.contains(pnt):
            b = True
            points.append(pnt)
            
```

```python
cod = []
for p in points:
    bef = len(cod)
    for j in inter_shp.index:
        if (inter_shp.geometry[j].contains(p)):
            cod.append(inter_shp.CD_RGINT[j])
            break
    aft = len(cod)
    if(bef == aft):
        coord = str(p.y) + ", " + str(p.x)
        prop = str(geolocator.reverse(coord))
        substring = "Região Geográfica Intermediária de "
        if (substring in prop):
            name = prop.split(substring)[1]
            name = name.split(",", 1)[0]
            cod.append(inter_shp[inter_shp['NM_RGINT'] == name].CD_RGINT)
```

```python
reg_interm = pd.DataFrame()
reg_interm['CD_RGINT'] = cod
reg_interm['GEOCODIGO'] = mun['code_muni']
reg_interm = pd.merge(reg_interm, pop, on = "GEOCODIGO")
reg_interm = reg_interm.groupby(["CD_RGINT"]).sum()
reg_interm = reg_interm.iloc[:, [3]]
reg_interm.to_csv("results/RegioesIntermediarias.csv", sep = ";")
```

```python
#Regioes Imediatas
ime_shp = geobr.read_immediate_region()
ime_shp
```

```python
cod = []
for p in points:
    for j in ime_shp.index:
        if (ime_shp.geometry[j].contains(p)):
            cod.append(int(ime_shp.code_immediate[j]))
            break
```

```python
reg_ime = pd.DataFrame()
reg_ime['CD_RGIME'] = cod
reg_ime['GEOCODIGO'] = mun['code_muni']
reg_ime = pd.merge(reg_ime, pop, on = "GEOCODIGO")
reg_ime = reg_ime.groupby(["CD_RGIME"]).sum()
reg_ime = reg_ime.iloc[:, [3]]
reg_ime.to_csv("results/RegioesImediatas.csv", sep = ";")
```
